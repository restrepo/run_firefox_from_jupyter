Full example:

````python
from pyvirtualdisplay import Display
from selenium import webdriver

display = Display(visible=0, size=(800, 600))
display.start()

browser = webdriver.Firefox()
browser.get(url)
html = browser.page_source

browser.close()
kk=display.stop()
````
