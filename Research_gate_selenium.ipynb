````python
import getpass
email=input('Research gate email: ')
password = getpass.getpass('Password: ')
````

````python
from pyvirtualdisplay import Display
from selenium import webdriver

display = Display(visible=0, size=(800, 600))
display.start()

browser = webdriver.Firefox()
````

````python
browser.get ("https://www.researchgate.net/application.Login.html")
````

````python
browser.find_element_by_id("input-login").send_keys(email)
````

````python
browser.find_element_by_id("input-password").send_keys(password)
````

````python
browser.find_element_by_class_name("nova-c-button__label").click()
````
